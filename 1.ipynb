```python
#2019091676 경영학부 장진하
#아래에서 사용할 random 모듈과 gcd, mod_inverse는 가독성을 위해 따로 뺐습니다.

import random

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def mod_inverse(a, m):
    (t_0, t_1) = (0, 1) 
    (r_0, r_1) = (m, a)

    while r_1 != 0:
        q = r_0 // r_1
        (t_0, t_1) = (t_1, t_0 - q * t_1) 
        (r_0, r_1) = (r_1, r_0 - q * r_1)

    if t_0 < 0 :
        t_0 += m

    return t_0
```

```python
def MR_PT(n, tol):
    #짝수 선별 제거
    if n % 2 == 0:
        return False
    t = n - 1
    
    a = random.randint(2,n-1)
    
    while t % 2 == 0:
        t = t // 2
    
    b = pow(a,t,n)
    '''
    MR primality test 'tol' times
    and return primeness of input 'n'
    '''
    if (b == 1) or (b == (n-1)):
        return True
        
    for i in range(1,tol):
        b = pow(b,2,n)
        
        if b == 1:
            return False
        elif b == (n-1):
            return True
            
    return False
```

```python
def RSA_setup(bit_length_for_primes):
    '''
    1. Generate two random primes p and q of length 'bit length for primes'
    2. Set N=p1
    3. Generate e and d satisfying ed = 1 (mod phi(N))
    4. return pk, sk, p, and q
    '''
    
    #p, q 생성
    for i in range(2):
        while True:
            n = random.getrandbits(bit_length_for_primes)
            if MR_PT(n, 100):
                if i == 0:
                    p = n
                    break
                else: #i == 1:
                    q = n
                    break
    
    #N 생성
    N = p * q
    #e 생성
    phi = (p-1) * (q-1)
    
    e = random.randint(2, phi - 1)
    while gcd(e,phi) != 1:
        e = random.randint(2, phi - 1)
    
    #d 생성
    d = mod_inverse(e, phi)
    
    #pk, sk 생성
    pk = [N, e]
    sk = [N, d]
    return pk, sk, p, q
```

```python
def RSA_EncDec (spk,MC) :
    #  = MC ^ (d,e) (mod N)
    MC = pow(MC, spk[1], spk[0])
    
    return MC
```
